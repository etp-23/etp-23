```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# Load the CSV file
file_path = 'first_1000.csv'
df = pd.read_csv(file_path)

# Assuming your comments are in a column named 'comments'
comments = df['text']

# Text preprocessing
# You can customize this based on your needs, e.g., removing punctuation and converting to lowercase
# For simplicity, let's just convert text to lowercase
comments = comments.str.lower()

from sklearn.feature_extraction.text import CountVectorizer

# create the vocabulary
vectorizer = CountVectorizer()

# fit the vocabulary to the text data
vectorizer.fit(comments)

# create the bag-of-words model
bow_model = vectorizer.transform(comments)

# print the bag-of-words model
print(bow_model)

# print the vocabulary
print(vectorizer.vocabulary_)
```
